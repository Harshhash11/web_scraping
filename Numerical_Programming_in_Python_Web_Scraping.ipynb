## **Scrapping Movies Data**

````python
def fetch_movie_urls(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
    }
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        return "Failed to retrieve the page, status code:", response.status_code
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup


url = 'https://www.justwatch.com/in/movies?release_year_from=2000'
soup=fetch_movie_urls(url)
print(soup.prettify())

## Hint : Use the following code to extract the film urls
# movie_links = soup.find_all('a', href=True)
# movie_urls = [link['href'] for link in movie_links if '/movie/' in link['href']]

# url_list=[]
# for x in movie_urls:
#   url_list.append('https://www.justwatch.com'+x)
````

## **Fetching Movie URL's**

````python
# Write Your Code here
movie_links = soup.find_all('a', href=True)
movie_urls = [link['href'] for link in movie_links if '/movie/' in link['href']]

url_list=[]
for x in movie_urls:
  url_list.append('https://www.justwatch.com'+x)

print(url_list) # Print the list of URLs to see the results
````

````python
import requests
from bs4 import BeautifulSoup
import re

movie_data = []  # List to hold all movie info dicts

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}

for movie_url in url_list:
    try:
        response = requests.get(movie_url, headers=headers)
        if response.status_code == 200:
            movie_soup = BeautifulSoup(response.text, 'html.parser')

            # Title and Release Year
            details_parent = movie_soup.find('div', class_='title-detail-hero__details')
            title = None
            release_year = None
            if details_parent:
                title_element = details_parent.find('h1', class_='title-detail-hero__details__title has-original-title')
                if title_element:
                    title = ''.join(t for t in title_element.find_all(string=True, recursive=False)).strip()
                year_element = details_parent.find('span', class_='release-year')
                release_year = year_element.get_text(strip=True).strip('()') if year_element else None

            # Director
            director = None
            director_section = movie_soup.find('h3', string='Director')
            if director_section:
                director_value = director_section.find_next_sibling('div', class_='poster-detail-infos__value')
                if director_value:
                    director_name_span = director_value.find('span', class_='title-credit-name')
                    if director_name_span:
                        director = director_name_span.get_text(strip=True)

            # Genres
            genres = []
            genre_section = movie_soup.find('h3', string='Genres')
            if genre_section:
                genre_value = genre_section.find_next_sibling('div', class_='poster-detail-infos__value')
                if genre_value:
                    span_text = genre_value.find('span')
                    if span_text:
                        genres_text = span_text.get_text(strip=True)
                        genres = [g.strip() for g in genres_text.split(',') if g.strip()]
                    linked_genres = genre_value.find_all('a', class_='detail-infos__value detail-infos__link')
                    for lg in linked_genres:
                        linked_genre = lg.get_text(strip=True)
                        if linked_genre and linked_genre not in genres:
                            genres.append(linked_genre)

            # IMDb Rating
            imdb_rating = None
            rating_section = movie_soup.find('h3', string='Rating')
            if rating_section:
                rating_value = rating_section.find_next_sibling('div', class_='poster-detail-infos__value')
                if rating_value:
                    imdb_div = rating_value.find('div', class_='jw-scoring-listing__rating--link')
                    if imdb_div:
                        imdb_text = imdb_div.find('div', class_='nowrap')
                        if imdb_text:
                            imdb_text_raw = imdb_text.get_text(strip=True)
                            match = re.search(r'(\d+(\.\d+)?)', imdb_text_raw)
                            imdb_rating = match.group(1) if match else None

            # Runtime
            runtime = None
            runtime_section = movie_soup.find('h3', string='Runtime')
            if runtime_section:
                runtime_value = runtime_section.find_next_sibling('div', class_='poster-detail-infos__value')
                if runtime_value:
                    runtime = runtime_value.get_text(strip=True)

            # Age Rating
            age_rating = None
            age_section = movie_soup.find('h3', string='Age rating')
            if age_section:
                age_value = age_section.find_next_sibling('div', class_='poster-detail-infos__value')
                if age_value:
                    age_rating = age_value.get_text(strip=True)

            # Production Countries
            prod_countries = None
            prod_section = movie_soup.find('h3', string='Production country')
            if prod_section:
                prod_value = prod_section.find_next_sibling('div', class_='poster-detail-infos__value')
                if prod_value:
                    prod_countries = prod_value.get_text(strip=True)

            # Streaming Services
            streaming_services = None
            streaming_section = movie_soup.select_one('section.spinning-texts div.first-paragraph p')
            if streaming_section:
                streaming_services = streaming_section.get_text(strip=True)

            # Append all info as dict
            movie_data.append({
                'Title': title,
                'Release Year': release_year,
                'Director': director,
                'Genres': genres,
                'IMDb Rating': imdb_rating,
                'Runtime': runtime,
                'Age Rating': age_rating,
                'Production Countries': prod_countries,
                'Streaming Services': streaming_services,
                'URL': movie_url
            })

        else:
            movie_data.append({
                'Title': f"Failed to retrieve page, status code: {response.status_code}",
                'Release Year': None,
                'Director': None,
                'Genres': [],
                'IMDb Rating': None,
                'Runtime': None,
                'Age Rating': None,
                'Production Countries': None,
                'Streaming Services': None,
                'URL': movie_url
            })

    except requests.exceptions.RequestException as e:
        movie_data.append({
            'Title': f"Error fetching URL: {e}",
            'Release Year': None,
            'Director': None,
            'Genres': [],
            'IMDb Rating': None,
            'Runtime': None,
            'Age Rating': None,
            'Production Countries': None,
            'Streaming Services': None,
            'URL': movie_url
        })

# Preview the first few entries
print(movie_data[:5])
````

## **Fetching Tv shows Url details**

````python
import requests
from bs4 import BeautifulSoup

# JustWatch India TV shows URL (from year 2000 onwards)
tv_url = 'https://www.justwatch.com/in/tv-shows?release_year_from=2000'

# Add headers to avoid getting blocked
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 '
                  '(KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9'
}

# Send request with headers
page = requests.get(tv_url, headers=headers)

tv_show_urls = []

if page.status_code == 200:
    soup1 = BeautifulSoup(page.text, 'html.parser')

    # Find all links
    tv_show_links = soup1.find_all('a', href=True)

    # Filter links to find TV show URLs
    # Assuming TV show links contain '/tv-show/'
    for link in tv_show_links:
        href = link['href']
        if '/tv-show/' in href:
            # Construct the full URL
            full_url = 'https://www.justwatch.com' + href
            tv_show_urls.append(full_url)

    # Print the list of TV show URLs (optional)
    print(tv_show_urls)
else:
    print(f"Failed to retrieve page: {page.status_code}")

# Now you have a list of TV show URLs in tv_show_urls
````

## **Creating TV Show DataFrame**

````python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
}

base_url = 'https://www.justwatch.com'
tv_show_urls = []

# Step 1: Get list of TV show URLs from multiple pages
for page in range(1, 6):  # 5 pages x ~20 TV shows per page = ~100
    tv_url = f'{base_url}/in/tv-shows?release_year_from=2000&page={page}'
    response = requests.get(tv_url, headers=headers)
    time.sleep(1)  # be polite

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        links = soup.find_all('a', href=True)
        for link in links:
            href = link['href']
            if '/tv-show/' in href and href not in tv_show_urls:
                tv_show_urls.append(base_url + href)
    else:
        print(f"Failed to retrieve page {page}: {response.status_code}")

print(f"Total TV Shows Found: {len(tv_show_urls)}")

# Step 2: Loop through TV show URLs and extract details
tv_show_data = []

for url in tv_show_urls[:60]:  # scrape first 60 to be safe
    time.sleep(1)
    try:
        res = requests.get(url, headers=headers)
        if res.status_code != 200:
            continue
        soup = BeautifulSoup(res.text, 'html.parser')

        # Title and Release Year
        title_tag = soup.find('h1', class_='title-detail-hero__details__title')
        if title_tag:
            title = ''.join(title_tag.find_all(string=True, recursive=False)).strip()
            year_tag = title_tag.find('span', class_='release-year')
            release_year = year_tag.text.strip('()') if year_tag else None
        else:
            title = release_year = None

        # Genres
        genres = []
        genre_block = soup.find('h3', string='Genres')
        if genre_block:
            genre_value = genre_block.find_next('div', class_='poster-detail-infos__value')
            if genre_value:
                genres = [g.strip() for g in genre_value.text.replace('\xa0', '').split(',') if g.strip()]

        # IMDb Rating
        imdb_rating = None
        imdb_div = soup.select_one('.jw-scoring-listing__rating--link .nowrap')
        if imdb_div:
            imdb_rating = imdb_div.get_text(strip=True).split()[0]

        # Age Rating
        age_rating = None
        age_heading = soup.find('h3', string='Age rating')
        if age_heading:
            age_div = age_heading.find_next('div', class_='poster-detail-infos__value')
            if age_div:
                age_rating = age_div.text.strip()

        # Production Countries
        countries = None
        country_heading = soup.find('h3', string='Production country')
        if country_heading:
            country_div = country_heading.find_next('div', class_='poster-detail-infos__value')
            if country_div:
                countries = country_div.text.strip()

        # Runtime
        runtime = None
        runtime_heading = soup.find('h3', string='Runtime')
        if runtime_heading:
            runtime_div = runtime_heading.find_next('div', class_='poster-detail-infos__value')
            if runtime_div:
                runtime = runtime_div.text.strip()

        # Streaming Service
        streaming_service = None
        stream_section = soup.find('section', class_='spinning-texts')
        if stream_section:
            paragraph = stream_section.find('p')
            if paragraph:
                streaming_service = paragraph.get_text(strip=True).replace('Currently you are able to watch "', '').replace('" streaming on ', '').replace('.', '')

        # Append all scraped data
        tv_show_data.append({
            'Title': title,
            'Release Year': release_year,
            'Genres': genres,
            'IMDb Rating': imdb_rating,
            'Age Rating': age_rating,
            'Production Countries': countries,
            'Runtime': runtime,
            'Streaming Service': streaming_service,
            'URL': url
        })

    except Exception as e:
        print(f"Error scraping {url}: {e}")
        continue

# Step 3: Create DataFrame
tv_df = pd.DataFrame(tv_show_data)
print(tv_df.head())
print(f"\nTotal TV Shows Scraped: {len(tv_df)}")
````

## **Finding Predominant Streaming Service**

````python
# Write Your Code here
import pandas as pd


# Combine both DataFrames if you want overall analysis
combined_df = pd.concat([movies_df, tv_df], ignore_index=True)

# Drop missing values in 'Streaming Services'
streaming_series = combined_df['Streaming Services'].dropna()

# Normalize the strings (some rows may say things like "Currently you are able to watch on Netflix" etc.)
# We'll extract the first service mentioned from the string
streaming_services_cleaned = streaming_series.str.extract(r'watch\s+"[^"]+"\s+streaming on ([^.,]+)', expand=False)
streaming_services_cleaned = streaming_services_cleaned.dropna().str.strip()

# Count occurrences
streaming_counts = streaming_services_cleaned.value_counts()

# Show top streaming service
print("📡 Predominant Streaming Service:")
print(streaming_counts.head(1))

# Optional: Show top 5
print("\nTop 5 Streaming Services:")
print(streaming_counts.head(5))
````

````python
#Let's Visvalize it using word cloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Join all streaming service names into one long string
text_data = ' '.join(streaming_services_cleaned)

# Create and configure word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='viridis',
    max_words=50,
    collocations=False
).generate(text_data)

# Plotting
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Predominant Streaming Services (Word Cloud)", fontsize=16)
plt.tight_layout()
plt.show()
````
